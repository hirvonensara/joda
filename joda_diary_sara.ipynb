Videon pääpointti ei kuitenkaan ollut varsinaisesti datan sisältö, vaan esitystapa. Rosling näytti konkreettisen esimerkkien kautta, millä tavalla dataa kannattaa esittää, jotta se on ymmärrettävää. Ensimmäisellä luennolla otettiin katsaus siihen, mitä datatiede on, mitä asioita siihen kuuluu ja millaisin työvälinein datatiedettä tehdään. Datatiedettä käsitteenä on vaikea määrittää hyvin lyhyesti, sillä se on monen eri asian summa. Datatiede voidaan jakaa neljään eri kokonaisuuteen, joita ovat **liiketoimintaosaaminen, ohjelmointiosaaminen, tilastollinen analyysi ja datalähtöinen viestintä ja visualisointi**. Liiketoimintaosaaminen on tärkeää, sillä datatieteen (tai yksittäisen tietotuotteen) tulisi lähteä liikkeelle jostakin tarpeesta, usein juuri liiketoiminnan tarpeesta. Tämä on mielestäni tärkeä huomio, sillä se mihin tarkoitukseen dataa hyödynnetään, määrittää sen, millä tavalla sitä analysoidaan. Itse analyysin tekemisessä korostuu tilastollisen analyysin menetelmien osaaminen, sekä usein ohjelmointiosaaminen. Nykyään datatiedettä voidaan kyllä tehdä myös erilaisten graafisten käyttöliittymien, jolloin ohjelmointiosaaminen ei ole välttämätöntä. Käytännössä edes jonkinlaisesta ohjelmointitaustasta on hyötyä. Datalähtöinen viestintä ja visualisointi taas liittyy samaan asiaan kuin Roslingin video, eli siihen, miten asiat esitetään. Tämä on se osa-alue, jonka tärkeys on ehkä korostanut viime vuosina eniten, sillä se on osa vähän suurempaa trendiä, eli ihmiskeskeistä suunnittelua. Ajatuksena on siis se, että siirrytään suunnittelemaan asioita helpommin ihmismielelle ymmärrettäväksi.
