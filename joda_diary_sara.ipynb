Datatiede voidaan jakaa neljään eri kokonaisuuteen, joita ovat **liiketoimintaosaaminen, ohjelmointiosaaminen, tilastollinen analyysi ja datalähtöinen viestintä ja visualisointi**. Itse analyysin tekemisessä korostuu tilastollisen analyysin menetelmien osaaminen, sekä usein ohjelmointiosaaminen. Tämä on se osa-alue, jonka tärkeys on ehkä korostanut viime vuosina eniten, sillä se on osa vähän suurempaa trendiä, eli ihmiskeskeistä suunnittelua. ### Kymmenen koodirivin demo

Ensimmäisen viikon demossa harjoitellaan pandasin käyttöä Kagglesta ladatulla Netflix-datalla. (https://www.kaggle.com/shivamb/netflix-shows) Datasetti on aika iso, joten tässä käytetään 150 ensimmäistä riviä. 

```python
import pandas as pd 
url = 'https://github.com/hirvonensara/joda/raw/master/demodata/netflix_titles.csv'
orig_df = pd.read_csv(url, nrows=150)
df = orig_df.copy()
print(df.columns)
titles_and_countries = df[['title','country']]
titles_by_countries = titles_and_countries.groupby(["country"]).sum
#print(titles_and_countries.tail())
print(titles_by_countries.tail())
```

Katsoin luennon ja koodiklinikan demoluennon (jälkikäteen). Luennolla keskusteltiin muun muassa datatieteen prosessista, liiketoimintarelevanssista ja menetelmistä. Lyhyesti ehdittiin käydä datatyyppejä ja datan hallintaa. 
Datatieteen prosessiin kuuluvia vaiheita on muun muassa esikäsittely, analyysi, reflektio ja tulosten viestiminen. Tärkeää prosessissa on huomata se, että analyysi ja reflektio käyvät vuoropuhelua, eli analyysi tehdään iteratiivisesti niin, että sen tuloksia arvioidaan ennen kuin päätetään "lopulliset" tulokset. 
Reflektion perusteella voidaan tutkia vaihtoehtoja. 
Datatieteen prosesseista voidaan erotella vielä erikseen kaksi prosessia, joita ovat ETL (Extract/Load/Transform) ja DAD (Discover/Access/Distill). Vincent Granvillen mukaan näistä ETL:ää käyttäisivät datainsinöörit ja DAD sen sijaan on datatieteen työprosessi. Nämä kaksi eroavat toisistaan siten, että ETL:ssä oletuksena on, että data on valmiiksi saatavilla jossain tietokannassa, laadukkaassa muodossa ja pitkällä aikajänteellä pysyvämpää. Sen sijaan DAD-prosessissa aikajänne on nopeampi ja prosessiin kuuluu myös datan kerääminen. Siihen kuuluu myös datan siivoaminen. 

Luennon yksi aihepiiri oli se, miten perustellaan data-analytiikan tärkeys/relevanssi liiketoiminnalle.
