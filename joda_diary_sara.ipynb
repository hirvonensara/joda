## 2. viikko

Katsoin luennon ja koodiklinikan demoluennon (jälkikäteen). Luennolla keskusteltiin muun muassa datatieteen prosessista, liiketoimintarelevanssista ja menetelmistä. Lyhyesti ehdittiin käydä datatyyppejä ja datan hallintaa. 
Datatieteen prosessiin kuuluvia vaiheita on muun muassa esikäsittely, analyysi, reflektio ja tulosten viestiminen. Tärkeää prosessissa on huomata se, että analyysi ja reflektio käyvät vuoropuhelua, eli analyysi tehdään iteratiivisesti niin, että sen tuloksia arvioidaan ennen kuin päätetään "lopulliset" tulokset. 
Reflektion perusteella voidaan tutkia vaihtoehtoja ja jopa palata prosessin alkuun määrittelemään kysymyksen asettelu uudestaan. Sama ilmiö voidaan havaita kyllä kaikessa tutkimuksessa: kun opitaan tutkittavasta asiasta lisää, voidaan huomata, että jokin toinen ongelma tai kysymys voi olla parempi kuin alkuperäinen.

Datatieteen prosesseista voidaan erotella vielä erikseen kaksi prosessia, joita ovat ETL (Extract/Load/Transform) ja DAD (Discover/Access/Distill). Vincent Granvillen mukaan näistä ETL:ää käyttäisivät datainsinöörit ja DAD sen sijaan on datatieteilijöiden työprosessi. Nämä kaksi eroavat toisistaan siten, että ETL:ssä oletuksena on, että data on valmiiksi saatavilla jossain tietokannassa, laadukkaassa muodossa ja pitkällä aikajänteellä pysyvämpää. Sen sijaan DAD-prosessissa aikajänne on nopeampi ja prosessiin kuuluu myös datan kerääminen. Siihen kuuluu myös datan siivoaminen. 

Luennon yksi aihepiiri oli se, miten perustellaan data-analytiikan tärkeys/relevanssi liiketoiminnalle. Data-analytiikka on tällä hetkellä aika trendikästä, ja monet organisaatiot haluavat pysyä ajassa kiinni ja tehdä data-analytiikkaa, vaikka eivät ymmärtäisi sen merkitystä omassa toiminnassaan. Banejreen et al. (?) artikkelin mukaan joissakin organisaatioissa ei ymmärretä data-analytiikan arvoa ja toisissa sitä taas pidetään ihmeenä, joka korjaa kaikki ongelmat. Kumpikaan ei kuitenkaan pidä paikkaansa, vaan data-analytiikka on hyödyllistä, jos sitä osataan tehdä oikein ja oikeaan tarpeeseen.
Eri tarpeisiin sopii erilainen analytiikan muoto, jä näitä muotoja ovat kuvaileva analytiikka, diagnosoiva analytiikka, ennakoiva analytiikka ja ohjaava analytiikka. Erityylisiä analytiikan keinoja tarvitaan eri tilanteissa. Nykyään tehdään paljon diagnosoivaa ja kuvailevaa analytiikkaa, mutta toivottavasti tulevaisuudessa voidaan siirtyä proaktiivisempaan analytiikkaan, kuten prediktiiviseen (ennustavaan) ja preskriptiiviseen (ohjaavaan). Toisaalta melkein aina on välttämätöntä tehdä jonkinlaista kuvailevaa analytiikkaa, jotta pystytään paremmin ymmärtämään, mitä kaikkea dataan liittyy ja mitä sillä voidaan tehdä.

Luennolla puhuttiin lopuksi vielä menetelmistä, joista ensimmäisenä datan keräämisen menetelmistä, joihin esimerkiksi ryömijät ja raapijat kuuluvat. Ryömijällä tarkoitetaan työkalua, käytännössä robottia, joka automaattisesti lataa jonkin nettiosoitteen sisältämän datan, pystyy seuraamaan osoitteen sisältämiä hyperlinkkejä ja lataamaan myös niiden sisältämän datan. Raapijalla taas tarkoitetaan työkalua, joka kerää automaattisesti spesifiä tietoa jostain nettiosoitteesta. Usein näitä käytetään yhdessä siten, että ryömijä luo hakuavaruuden, josta tieto etsitään ja raapija kerää annettujen hakuparametrien perusteella spesifiä tietoa. Yksi tärkeimmistä datan hallinnan työkaluista on Pythonin Pandas-kirjasto. Pandas on vähän niin kuin Pythonin Excel, eli sitä käytetään suurien datamassojen hallintaan ohjelmallisesti. Pandasin etuja on sen tehokkuus ja helppokäyttöisyys.
